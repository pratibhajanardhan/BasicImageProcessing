```python
df['Word Count'] = df['feedback before cleaning'].apply(lambda x: len(str(x).split(" ")))
Data = df['feedback before cleaning'].str.replace('[\r\n]', '')
Data = pd.DataFrame(Data)
Data.head()

Data['char_count'] = Data['feedback before cleaning'].str.len() # This also include space
Data.head()

# No.of words
Data['word_count'] = Data['feedback before cleaning'].apply(lambda x: len(str(x).split(" ")))
Data.head()
```

```python
def avg_word(sentence):
  words = sentence.split()
  return (sum(len(word) for word in words)/len(words))

Data['avg_word'] = Data['feedback before cleaning'].apply(lambda x: avg_word(x))
Data.head()
```

```python
# No.of stop words
from nltk.corpus import stopwords
stop = stopwords.words('english')

Data['stopwords'] = Data['feedback before cleaning'].apply(lambda x: len([x for x in x.split() if x in stop]))
Data.head()
```

```python
# No.of special charecters
Data['hastags'] = Data['feedback before cleaning'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
Data.head()
```

```python
# Number of numerics
Data['numerics'] = Data['feedback before cleaning'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
Data.head()
```

```python
# No.of Uppercase words
Data['upper'] = Data['feedback before cleaning'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
Data.head()
```

```python
#The first pre-processing step which we will do is transform our tweets into lower case. 
#This avoids having multiple copies of the same words. For example, while calculating the word count, ‘Analytics’ and ‘analytics’ will be taken as different words.
Data['feedback before cleaning'] = Data['feedback before cleaning'].apply(lambda x: " ".join(x.lower() for x in x.split()))
Data.head()
```

```python
#As we discussed earlier, stop words (or commonly occurring words) should be removed from the text data. 
#For this purpose, we can either create a list of stopwords ourselves or we can use predefined libraries.
from nltk.corpus import stopwords
stop = stopwords.words('english')
Data['feedback before cleaning'] = Data['feedback before cleaning'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
Data.head()
```

```python
#Previously, we just removed commonly occurring words in a general sense.
#We can also remove commonly occurring words from our text data First, let’s check the 10 most frequently occurring words in our text data then take call to remove or retain.
freq = pd.Series(' '.join(Data['feedback before cleaning']).split()).value_counts()[:10]
freq
```

```python
# Now, let’s remove these words as their presence will not of any use in classification of our text data
freq = list(freq.index)
Data['feedback before cleaning'] = Data['feedback before cleaning'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
Data.head()
```

```python
#Similarly, just as we removed the most common words, this time let’s remove rarely occurring words from the text.
#Because they’re so rare, the association between them and other words is dominated by noise. You can replace rare words with a more general form and then this will have higher counts.
freq = pd.Series(' '.join(Data['feedback before cleaning']).split()).value_counts()[-10:]
freq
```

```python
freq = list(freq.index)
Data['feedback before cleaning'] = Data['feedback before cleaning'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
Data.head()
```
